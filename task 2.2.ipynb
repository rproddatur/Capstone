# BIVAIRIATE

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
plt.title('Duration to accept offer vs Notice period(r = {0:0.2f})'.format(pearsonr(drop['Duration to accept offer'],drop['Notice period'])[0]))
sns.scatterplot(x="Duration to accept offer",y="Notice period",data = df,hue="Status")
```

```python
plt.title('Duration to accept offer vs Notice Period(r = {0:0.2f})'.format(pearsonr(drop['Duration to accept offer'],drop['Notice period'])[0]))
sns.scatterplot(x="Duration to accept offer", y="Notice period",hue="Status", style="Status", data=df)
plt.show()
```

```python
plt.title('mba_p vs degree_p(r = {0:0.2f})'.format(pearsonr(drop['Duration to accept offer'],drop['Notice period'])[0]))
sns.scatterplot(x="Duration to accept offer", y="Notice period",hue="Status", style="Status", data=df)
plt.show()
```
