```python
# test the claim that the proportion of population of smoking is 40%
# H0: P = 0.4 & H1: P != 0.4 (One sample Z-proportion test)
# Z_statistic = (p-P)/(sqrt(p*(1-p)/n))
P = 0.4
alpha = 0.05
x = 93
n = len(df)
p = x/n
se = np.sqrt(p*(1-p)/n)
z_stat =(p-P)/se
zalphaby2 = st.norm.isf(alpha/2)
p_value = st.norm.sf(abs(z_stat))*2
print(z_stat,zalphaby2,p_value)
```
